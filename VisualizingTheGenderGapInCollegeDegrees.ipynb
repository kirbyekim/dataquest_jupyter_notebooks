```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

subcats = [stem_cats, lib_arts_cats, other_cats]

fig = plt.figure(figsize=(13, 23))
for col in range(1,4):
    for sp in range(0,6):
        ax = fig.add_subplot(6,3,(sp * 3) +col)
        if sp < len(subcats[col - 1]):
            ax.plot(women_degrees['Year'], women_degrees[subcats[col - 1][sp]], c=cb_dark_blue, label='Women', linewidth=3)
            ax.plot(women_degrees['Year'], 100-women_degrees[subcats[col - 1][sp]], c=cb_orange, label='Men', linewidth=3)
            ax.spines["right"].set_visible(False)    
            ax.spines["left"].set_visible(False)
            ax.spines["top"].set_visible(False)    
            ax.spines["bottom"].set_visible(False)
            ax.set_xlim(1968, 2011)
            ax.set_ylim(0,100)
            ax.set_yticks([0,100])
            ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
            ax.set_title(subcats[col - 1][sp])
            ax.tick_params(bottom="off", top="off", left="off", right="off")
            if (sp * 3) +col not in ([14,16,18]):
                #pass
                ax.set_xticks([])
            if sp == 0:
                if women_degrees[subcats[col - 1][sp]].iloc[-1] > 100 - women_degrees[subcats[col - 1][sp]].iloc[-1]:
                    ax.text(2005, women_degrees[subcats[col - 1][sp]].iloc[-1] + 12, 'Men')
                    ax.text(2002, 100 - women_degrees[subcats[col - 1][sp]].iloc[-1] - 15, 'Women')
                else:
                    ax.text(2005, women_degrees[subcats[col - 1][sp]].iloc[-1] - 15, 'Men')
                    ax.text(2002, 100 - women_degrees[subcats[col - 1][sp]].iloc[-1] + 12, 'Women')
            elif sp == len(subcats[col - 1]) -1:
                if women_degrees[subcats[col - 1][sp]].iloc[-1] > 100 - women_degrees[subcats[col - 1][sp]].iloc[-1]:
                    ax.text(2005, women_degrees[subcats[col - 1][sp]].iloc[-1] + 12, 'Men')
                    ax.text(2002, 100 - women_degrees[subcats[col - 1][sp]].iloc[-1] - 15, 'Women')
                else:
                    ax.text(2005, women_degrees[subcats[col - 1][sp]].iloc[-1] - 15, 'Men')
                    ax.text(2002, 100 - women_degrees[subcats[col - 1][sp]].iloc[-1] + 12, 'Women')
plt.show()
plt.savefig('gender_degrees.png')
```
