```python
import pandas as pd
import numpy as np
```

```python
#read csv data contained in thanksgiving.csv. The dataset came from 
#FiveThirtyEight
data = pd.read_csv('thanksgiving.csv', encoding="Latin-1")
```

```python
#show counts of how many do or do not celebrate
a = data['Do you celebrate Thanksgiving?']
a.value_counts()
```

```python
#replace data DF with just those who DO celebrate Tahnksgiving
data = data[data['Do you celebrate Thanksgiving?'] == 'Yes']
```

```python
#show distinct values and counts for what people eat for thanksgiving 
#main dish
main = data['What is typically the main dish at your Thanksgiving dinner?']
main.value_counts()
```

```python
#filter for just tofurkey eaters and show which ones have gravy
tofurkey_eaters = data[data['What is typically the main dish at your Thanksgiving dinner?'] == 'Tofurkey']
tofurkey_eaters['Do you typically have gravy?']
```

```python
#filter for those that don't eat applie pie, don't eat pumpkin pie 
#and dont eat pecan pie
apple_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple'])
pumpkin_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin'])
pecan_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan'])

#show counts of who doesn't eat any of the 3 types of pie and who 
#does
ate_pies = apple_isnull & pumpkin_isnull & pecan_isnull
ate_pies.value_counts()
```

```python
#function to tranform Age to strip + sign and get first Age in Range 
def convert_toint(str):
    if pd.isnull(str):
        return None
    parts = str.split(' ')
    age = int(parts[0].replace('+',''))
    return age
```

```python
#apply convert_toint to Age column
data["int_age"] = data["Age"].apply(convert_toint)
```

```python
#function to get an integer income from the 'How much total combined 
#money did all members of your HOUSEHOLD earn last year?' column
import re
def get_income(str):
    if pd.isnull(str):
        return None
    parts = str.split(' ')
    if re.match(r'^Prefer', parts[0]):
        return None
    income = parts[0].replace(',','')
    income = int(income.replace('$',''))
    return income
data["int_income"] = data["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(get_income)
```

```python
#filter int_income column < $150,000
inc_lt_15K = data[data['int_income'] < 150000]
inc_lt_15K.head(5)
```

```python
#filter int_income column > $150,000
inc_gt_15K = data[data['int_income'] > 150000]
inc_gt_15K.head(5)
```

```python
#show distinct  counts for 'How far will you travel for 
#Thanksgiving?' colunm in Series containing people who make more 
#than $150,000
inc_gt_15K['How far will you travel for Thanksgiving?'].value_counts()
```

```python
#pivot table showing avg age by "Have you ever tried to meet up with hometown friends on Thanksgiving night?' 
#grouped by 'Have you ever attended a "Friendsgiving?"'
data.pivot_table(index=['Have you ever tried to meet up with hometown friends on Thanksgiving night?'],values=['int_age'], columns=['Have you ever attended a "Friendsgiving?"'])
```

```python
#pivot table showing av income by "Have you ever tried to meet up with hometown friends on Thanksgiving night?' 
data.pivot_table(index=['Have you ever tried to meet up with hometown friends on Thanksgiving night?','Have you ever attended a "Friendsgiving?"'],values=['int_income'])
```
